```python
import pandas as pd
import numpy as np
import os
import tensorflow as tf
from tensorflow.keras import layers

DATASET = os.path.join("data", "training.csv")
csv = pd.read_csv(DATASET)
```

```python
dataset = csv.to_numpy()
```

```python
sample_size = 15
aux_Y = dataset[:, -1].astype(np.int)
aux_X = dataset[:, 1:-1]
Y = []
X = []
for i in range(0, dataset.shape[0], sample_size):
    interval = slice(i, i + sample_size)
    if ((aux_Y[interval] == 0).all() or (aux_Y[interval] == 1).all()) and len(aux_Y[interval]) == sample_size:
        Y.append(aux_Y[i])
        X.append(aux_X[interval])        
```

```python
model = tf.keras.Sequential()
model.add(layers.LSTM(128))
model.add(layers.Dense(2, activation='softmax'))
model.compile(loss='categorical_crossentropy', optimizer='Adam', metrics=['accuracy'])
```

```python
early_stopping_callback = tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=40)
checkpoint_callback = tf.keras.callbacks.ModelCheckpoint("lstm_checkpoint.h5", monitor='val_loss', verbose=1, save_best_only=True, mode='min')
history = model.fit(
    np.array(X),
    np.array(Y),
    validation_split = 0.2,
    epochs = 40,
    batch_size = 18,
    callbacks=[early_stopping_callback, checkpoint_callback]
)
```
